###MEDIDAS DE DISPERSÃO

```python
# (* Medidas de Dispersão *) é como e o comportamento da distribuição dos dados dentro do conjuno  de dados
```

```python
# ( * Amplitude * ) é a diferença entre o menor e o maior dentro do conjuto de dados
df.NOTA_CN.max() - df.NOTA_CN.min() # o maximo - min de um conjunto de dados
```

```python
# ( * Variancia * ) ou variação dentro do conjunto de dados
df.NOTA_CN.var()
```

```python
# ( * Desvio Padrão *)   ou    std
df.NOTA_CN.std()
```
